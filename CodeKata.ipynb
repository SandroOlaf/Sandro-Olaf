### **Exerc√≠cio 1: O "Gritador"** O que voc√™ vai praticar: Passar uma fun√ß√£o simples para outra.

```python

# Neste exerc√≠cio, voc√™ vai criar uma fun√ß√£o que decide como uma mensagem deve ser exibida.

# Crie duas fun√ß√µes oper√°rias:

# simple_print(text): apenas d√° um print(text).
def simple_print(text):
  print(text)
# decorated_print(text): d√° um print com estrelinhas em volta, tipo *** text ***.
def decorated_print(text):
  print(f"*** {text} ***")

# Crie a Higher-Order Function:
# Higher-Order Function √© apenas uma fun√ß√£o "chefe" que pede para uma fun√ß√£o "oper√°ria" fazer uma parte do trabalho.

# display_message(message, formatter_function).
def display_message(message, formatter_function):
  formatter_function(message)

display_message('Hello', simple_print)
display_message('Hello', decorated_print)

# A display_message deve apenas chamar a fun√ß√£o que voc√™ passou para ela.

```

### **Exerc√≠cio 2: O Operador de N√∫meros** # O que voc√™ vai praticar: Retornar valores de uma fun√ß√£o oper√°ria para a fun√ß√£o chefe.

```python
# Vamos criar uma fun√ß√£o que sabe lidar com n√∫meros, mas n√£o sabe qual conta fazer at√© voc√™ mandar.

# Crie as fun√ß√µes: double(n) (retorna n * 2) e triple(n) (retorna n * 3).
def double(n):
  return n * 2
def triple(n):
  return n * 3
# Crie a fun√ß√£o chefe: apply_operation(number, operation_function).
def apply_operation(number, operation_function):
  result = operation_function(number)
  print(f"O resultado da opera√ß√£o {apply_operation.__name__} √©: [{result}]")

apply_operation(2, double)
apply_operation(2, triple)


# Ela deve aplicar a opera√ß√£o ao n√∫mero e imprimir: "O resultado da opera√ß√£o [nome da fun√ß√£o] √©: [resultado]". (Dica: use .__name__).

```

### **Exerc√≠cio 3: O Filtro de Comidas** # O que voc√™ vai praticar: Usar Higher-Order Functions para filtragem (muito comum no dia a dia).

```python
# Imagine que voc√™ tem uma lista de frutas: ['ma√ß√£', 'banana', 'manga', 'uva', 'melancia'].

# Crie uma fun√ß√£o chefe chamada food_filter(list, criteria_function).
frutas = ['ma√ß√£', 'banana', 'manga', 'uva', 'melancia']

# Crie duas fun√ß√µes de crit√©rio:

# starts_with_m(word): retorna True se come√ßar com "m".
def starts_with_m(word):
  return word.startswith("m")

# short_word(word): retorna True se a palavra tiver menos de 5 letras.
def short_word(word):
  return len(word) > 5

# A fun√ß√£o chefe deve percorrer a lista e imprimir apenas as frutas que a fun√ß√£o de crit√©rio aprovar.

def food_filter(list_of_foods, criteria_function):
  print(f"\n--- Filtrando com a regra: {criteria_function.__name__} ---")
  for fruta in list_of_foods:
    if criteria_function(fruta):
      print(f"‚úÖ Aprovada: {fruta}")


food_filter(frutas, starts_with_m)
food_filter(frutas, short_word)
# O que voc√™ vai praticar: Usar Higher-Order Functions para filtragem (muito comum no dia a dia).
```

```python
# Imagine que voc√™ tem uma lista de pre√ßos de produtos: [10.50, 99.00, 45.00, 150.00, 20.00, 300.00].

prices =  [10.50, 99.00, 45.00, 150.00, 20.00, 300.00]

# Regra 1 (is_cheap): Retorna True se o pre√ßo for menor que 50.
def is_cheap(preco):
  return preco > 50

# Regra 2 (is_luxury): Retorna True se o pre√ßo for maior que 200.
def is_luxury(preco):
  return preco > 200

# Fun√ß√£o Chefe (price_analyzer): Percorre a lista e imprime: "Pre√ßo detectado: R$ [valor]" apenas para os que passarem no crit√©rio.

# def price_analyzer(produtos, retorno):
#   print(f"\n--- Filtrando com a regra: {retorno.__name__} ---")
#   for i in produtos:
#     if retorno(i):
#       print(f'Valor {i}')

def price_analyzer(price_list, criteria_function):
    # Usamos o .__name__ para deixar o print elegante
    print(f"\n--- Analisando: {criteria_function.__name__} ---")

    for p in price_list:
        # Aqui a m√°gica acontece: a HOF chama a regra para cada pre√ßo
        if criteria_function(p):
            print(f"Pre√ßo detectado: R$ {p:.2f}") # .2f formata para 2 casas decimais


# Teste 1: Procurando promo√ß√µes
price_analyzer(prices, is_cheap)

# Teste 2: Procurando itens exclusivos
price_analyzer(prices, is_luxury)
```

```python
contatos = ["ana@gmail.com", "admin@empresa.com", "suporte@empresa.com", "joao@outlook.com"]

def is_corporate(email):
  return email.endswith(".com")

def is_gmail(email):
  return email.endswith("@gmail.com")

def is_outlook(email):
  return email.endswith("@outlook.com")

def email_filter(email_list, criteria_function):
  print(f"\n=-=-=-= analisando: {criteria_function.__name__} ")
  for email in email_list:
    if criteria_function(email):
      print(email)

email_filter(contatos, is_corporate)
email_filter(contatos, is_gmail)
email_filter(contatos, is_outlook)

```

```python
temperaturas = [18, 25, 32, 10, 40, 22, 5,55]

def is_danger_hot(temp):
  return temp > 35

def is_danger_cold(temp):
  return temp < 10

def alert_system(temp_list, criteria_function):
  print(f"\n---- Temperaturas Detectadas: {criteria_function.__name__}")
  for tempo in temp_list:
    if criteria_function(tempo):
      print(f"{tempo}¬∞C")

alert_system(temperaturas, is_danger_hot)
alert_system(temperaturas, is_danger_cold)


```

```python
real = [100, 250, 50, 10]

def to_dollar(cash):
  return cash / 5.54

def to_euro(cash):
  return cash / 6.53

def convert_prices(convert_list, criteria_function):
  print(f"\n---Convertendo Valores: {criteria_function.__name__}")
  for converte in convert_list:
    convertido = criteria_function(converte)
    print(f"R$ {converte} -> Convertido: {convertido:.2f}")

convert_prices(real, to_dollar)
convert_prices(real, to_euro)
```

```python
# 1. O DECORATOR (A Capinha)
def meu_decorator(funcao_original):
    def wrapper():
        print("--- Log: Iniciando execu√ß√£o ---") # Comportamento extra antes
        funcao_original()                       # Executa a fun√ß√£o original
        print("--- Log: Finalizado com sucesso ---") # Comportamento extra depois
    return wrapper

# 2. USANDO O DECORATOR (@)
@meu_decorator
def saudar():
    print("Ol√°, eu sou uma fun√ß√£o simples!")

# 3. EXECUTANDO
saudar()
```

```python
aluno = {
    "nome": "Sandro Olaf",
    "curso": "ADS",
    "nota": 7
}

aluno['curso'] = "Python Master"

def verificar_aprova√ß√£o(dados_aluno):
  nome_do_aluno = dados_aluno["nome"]
  nota_do_aluno = dados_aluno["nota"]
  curso_do_aluno = dados_aluno["curso"]

  if nota_do_aluno >= 7:
    print(f"O aluno {nome_do_aluno} foi aprovado em {curso_do_aluno}!")
  else:
    print(f"O aluno {nome_do_aluno} precisa de recupera√ß√£o.")

verificar_aprova√ß√£o(aluno)
```

### **O Filtro de Estoque (HOF + Dicion√°rio)**

```python
estoque = [
    {"nome": "Mouse", "preco": 50},
    {"nome": "Monitor", "preco": 800},
    {"nome": "Fone", "preco": 200}
]

def filter_produtos(produto):
    return produto["preco"] >= 100

def verificar_item(lista, criteria_function):
    print(f"\n---Verificando Itens : {criteria_function.__name__}")
    for item in lista:
        # Aqui est√° a m√°gica: o 'if' chama a regra para cada item
        if criteria_function(item):
            print(f"Produto: {item['nome']} | R$ {item['preco']}")
verificar_item(estoque, filter_produtos)
```
